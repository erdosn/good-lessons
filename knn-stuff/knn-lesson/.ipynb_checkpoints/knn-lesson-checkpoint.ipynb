# KNearest Neighbors

What is goal of using KNearest Neighbors?

You're working as a data scientist for Facebook.   Maybe you want to predict the number of friends someone is going to have based on their metadata. 

Metadata:
    * Location
    * Likes per post
    * Gender
    * Age
    * Occupation
    * Political Affiliation
    * Groups you follow
    * Followers
    * Friends (initial)
    * IP addresses

Target:
    * Total number of friends added by 1 year
    
    
How would KNN solve this problem?
    * Looks at your Nearest Neighbors. 
        * Literally the people closest to you in our data space. 
        * It looks at their number of friends
        * Makes a prediction based on their numbers
    * Is this Regression or Classification (Target dependent)
        * Regression - because we're predicting number of friends (continuous variable)
        * Classification - predicting color, gender

````python
import pandas as pd
import numpy as np

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.model_selection import train_test_split

from sklearn.datasets import load_iris

import matplotlib.pyplot as plt
import seaborn as sns
````

````python
iris = load_iris()
data = iris.data
features = iris.feature_names
target = iris.target
target_names = iris.target_names
````

````python
df = pd.DataFrame(data, columns=features)
df['target'] = target
df['target_names'] = [target_names[i] for i in target]
df.head()
````

# Predicting the flower type is Classification - because we're predicting which category

````python
df['target_names'].value_counts()
````

````python
plt.figure(figsize=(10, 10))
sns.pairplot(data=df, hue='target_names', corner=True)
plt.show()
````

# Lets build a KNN and evaluate it

````python
df_train, df_test = train_test_split(df, test_size=0.30)
````

````python
df_train.head()
````

````python
df_test.head()
````

````python
df_train.shape, df_test.shape
````

````python
features_used = ['petal length (cm)', 'petal width (cm)']
````

````python
# We aren't turning the hyperparameters
knn = KNeighborsClassifier(n_neighbors=3, n_jobs=-1)
knn.fit(df_train[features_used], df_train['target_names'])
````

````python
# what does the fit method do in KNN?
sns.scatterplot(x='petal length (cm)', y='petal width (cm)', hue='target_names', data=df_train)
plt.show()
````

````python
random_point = df_test.sample(n=1)
````

````python
random_point
````

````python
random_point[features_used]
````

````python
# how do we predict a point using our knn model?
knn.predict(random_point[features_used])
````

````python
test_predictions = knn.predict(df_test[features_used])
test_predictions
````

````python
# so let's evaluate our model
# accuracy = (number correct )/ (total points attempted)
# accuracy = (TP + TN) / (TP + FP + TN + FN)
````

````python
# let's build a confusion matrix
````

````python
cm = confusion_matrix(df_test['target_names'], test_predictions, labels=target_names)
````

````python
target_names
````

````python
sns.heatmap(cm, annot=True, cmap=sns.color_palette('Blues'), 
            xticklabels=target_names, yticklabels=target_names)
plt.title("Confusion Matrix")
plt.xlabel("Predicted", fontsize=16)
plt.ylabel("Actual", fontsize=16)
plt.show()
````

````python
# model.score is always accuracy in sklearn
knn.score(df_test[features_used], df_test['target_names'])
````

````python
# how do we find the points we got incorrect?
````

````python
df_test['predicted_target'] = test_predictions
df_test.head()
````

````python
df_test_incorrect = df_test[df_test['target_names']!=df_test['predicted_target']]
````

````python
sns.scatterplot(x='petal length (cm)', y='petal width (cm)', hue='target_names', data=df_train)
sns.scatterplot(x='petal length (cm)', y='petal width (cm)', data=df_test_incorrect)

plt.show()
````

````python
df_test['correct'] = df_test['target_names']==df_test['predicted_target']
df_test.head()
````

````python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='petal length (cm)', y='petal width (cm)', hue='predicted_target', s=200,
                style='correct', data=df_test)
plt.show()
````
